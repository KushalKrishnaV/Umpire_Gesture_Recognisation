```python
from sklearn.model_selection import train_test_split

def load_images_and_labels_from_folder(folder):
    images = []
    labels = []
    for class_folder in os.listdir(folder):
        class_path = os.path.join(folder, class_folder)
        if os.path.isdir(class_path):
            for filename in os.listdir(class_path):
                # Load image and handle errors
                try:
                    img = Image.open(os.path.join(class_path, filename))
                    # Resize image if necessary
                    img = img.resize((224, 224))  # Adjust size as needed
                    images.append(np.array(img))  # Convert image to array
                    labels.append(class_folder)
                except Exception as e:
                    print(f"Error loading image: {e}")
    return images, labels

# Set your train and test data directories
train_dir = r"C:\Users\Kusha\Desktop\masters\final_sem\project\knn_rff\train_data"
test_dir = r"C:\Users\Kusha\Desktop\masters\final_sem\project\knn_rff\test_data"

# Load train images and labels
train_images, train_labels = load_images_and_labels_from_folder(train_dir)

# Load test images and labels
test_images, test_labels = load_images_and_labels_from_folder(test_dir)

# Convert train images to arrays
X_train = np.array(train_images)

# Convert test images to arrays
X_test = np.array(test_images)

# Convert labels to numerical values
label_to_index = {label: i for i, label in enumerate(set(train_labels))}
train_labels = np.array([label_to_index[label] for label in train_labels])
test_labels = np.array([label_to_index[label] for label in test_labels])

# Split data into train and test sets
X_train, X_val, y_train, y_val = train_test_split(X_train, train_labels, test_size=0.2, random_state=42)

# Now you have your train, validation, and test data split into arrays and labels.
# X_train, X_val, and X_test contain the images as arrays, while y_train, y_val, and y_test contain the corresponding labels.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Initialize Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the training data
rf_classifier.fit(X_train.reshape(len(X_train), -1), y_train)

```

```python
# Predict on the validation set
y_val_pred = rf_classifier.predict(X_val.reshape(len(X_val), -1))
y_val_pred
```

```python
# Calculate accuracy on the validation set
val_accuracy = accuracy_score(y_val, y_val_pred)
print("Validation Accuracy:", val_accuracy)
```

```python
# Predict on the test set
y_test_pred = rf_classifier.predict(X_test.reshape(len(X_test), -1))
```

```python
# Calculate accuracy on the test set
test_accuracy = accuracy_score(test_labels, y_test_pred)
print("Test Accuracy:", test_accuracy)
```
